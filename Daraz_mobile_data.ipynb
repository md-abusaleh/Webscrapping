```python
import requests

data = []

for page in range(1, 12):
    url = f"https://www.daraz.com.bd/smartphones/?ajax=true&page={page}"
    response = requests.get(url)
    data_json = response.json()
    print(f'{page} no page is done')

    all_phones = data_json["mods"]["listItems"]
    for item, phone in enumerate(all_phones):
        try:
            originalPrice = phone.get('originalPrice', phone['price'])
        except KeyError:
            originalPrice = phone['price']
        
        try:
            originalPriceShow = phone.get('originalPriceShow', phone['price'])
        except KeyError:
            originalPriceShow = phone['price']
        
        discount = phone.get('discount', 'Not Available')
        
        ratingScore = phone.get('ratingScore', "None")
        
        d = {
            'name': phone['name'],
            'originalPrice': originalPrice,
            'originalPriceShow': originalPriceShow,
            'price': phone['price'],
            'priceShow': phone['priceShow'],
            'discount': discount,
            'ratingScore': ratingScore,
            'brandName': phone['brandName'],
            'sellerName': phone['sellerName'],
        }
        data.append(d)
```
